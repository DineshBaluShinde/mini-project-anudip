```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("students.csv")
df
```

```python
df["Total"] = df[["Maths", "Science", "English"]].sum(axis=1)
df["Average"] = np.mean(df[["Maths", "Science", "English"]], axis=1)

df
```

```python
highest_marks = df["Total"].max()
lowest_marks = df["Total"].min()
average_marks = df["Total"].mean()

print("Highest Marks:", highest_marks)
print("Lowest Marks:", lowest_marks)
print("Average Marks:", average_marks)
```

```python
df["Result"] = np.where(df["Average"] >= 50, "Pass", "Fail")

pass_percentage = (df["Result"].value_counts(normalize=True)["Pass"]) * 100

print("Pass Percentage:", pass_percentage)
df
```

```python
plt.figure()
plt.bar(df["Name"], df["Total"])
plt.xlabel("Student Name")
plt.ylabel("Total Marks")
plt.title("Student Performance Analysis")
plt.show()
```

```python
df["Result"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Pass vs Fail Percentage")
plt.ylabel("")
plt.show()
```

```python
df.to_csv("student_result_analysis.csv", index=False)
print("Result file saved successfully!")
```
