#### Employee Attendance and Salary Analyzer: Create attendance data and salary dataset. Calculate attendance percentage, apply tax rules, and visualize salary distribution using histogram.Use Python, NumPy and Pandas.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Employee data
data = {
    "Employee_ID": [101, 102, 103, 104, 105],
    "Employee_Name": ["Amit", "Neha", "Ravi", "Pooja", "Kiran"],
    "Total_Working_Days": [26, 26, 26, 26, 26],
    "Days_Present": [24, 22, 20, 25, 18],
    "Basic_Salary": [30000, 35000, 28000, 40000, 25000]
}

df = pd.DataFrame(data)
df
```

```python
df["Attendance_Percentage"] = (df["Days_Present"] / df["Total_Working_Days"]) * 100
df
```

```python
def calculate_tax(salary):
    if salary <= 30000:
        return salary * 0.05
    elif salary <= 40000:
        return salary * 0.10
    else:
        return salary * 0.15

df["Tax"] = df["Basic_Salary"].apply(calculate_tax)
df["Net_Salary"] = df["Basic_Salary"] - df["Tax"]
df
```

```python
plt.figure()
plt.hist(df["Net_Salary"], bins=5)
plt.xlabel("Net Salary")
plt.ylabel("Number of Employees")
plt.title("Salary Distribution Histogram")
plt.show()
```
